## Q. select the survived male passenegers and all the people who boarded ship from port "S"

```python
condition1 = train['Embarked'] == 'S'
condition2 = train['Sex'] == 'male'
condition3 = train['Survived'] == 1

condition = (condition1 | (condition2 & condition3))
```

```python
train[condition]
```
